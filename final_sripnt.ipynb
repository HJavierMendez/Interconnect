# Proyecto final 

<h2> Descripcion del proyecto </h2>

Al operador de telecomunicaciones Interconnect le gustaría poder pronosticar su tasa de cancelación de clientes. Si se descubre que un usuario o usuaria planea irse, se le ofrecerán códigos promocionales y opciones de planes especiales. El equipo de marketing de Interconnect ha recopilado algunos de los datos personales de sus clientes, incluyendo información sobre sus planes y contratos.

<h2> Servicios de Interconnect</h2>

Interconnect proporciona principalmente dos tipos de servicios:

1. Comunicación por teléfono fijo. El teléfono se puede conectar a varias líneas de manera simultánea.
2. Internet. La red se puede configurar a través de una línea telefónica (DSL, *línea de abonado digital*) o a través de un cable de fibra óptica.

Algunos otros servicios que ofrece la empresa incluyen:

- Seguridad en Internet: software antivirus (*ProtecciónDeDispositivo*) y un bloqueador de sitios web maliciosos (*SeguridadEnLínea*).
- Una línea de soporte técnico (*SoporteTécnico*).
- Almacenamiento de archivos en la nube y backup de datos (*BackupOnline*).
- Streaming de TV (*StreamingTV*) y directorio de películas (*StreamingPelículas*)

La clientela puede elegir entre un pago mensual o firmar un contrato de 1 o 2 años. Puede utilizar varios métodos de pago y recibir una factura electrónica después de una transacción.

<h2> Descripción de los datos</h2>

Los datos consisten en archivos obtenidos de diferentes fuentes:

- `contract.csv` — información del contrato;
- `personal.csv` — datos personales del cliente;
- `internet.csv` — información sobre los servicios de Internet;
- `phone.csv` — información sobre los servicios telefónicos.

En cada archivo, la columna `customerID` (ID de cliente) contiene un código único asignado a cada cliente. La información del contrato es válida a partir del 1 de febrero de 2020.

#### 2.3.6 Obsvacion del analisis de los datos 

Se observa que todos los data frames tiene en comun la primer columnas de customer_id,tomando encuenta esto se realizara un merge, para mejorar el analisis

```python
info_all = personal.merge(phone, on='customer_id', how='outer')
info_all = info_all.merge(internet, on='customer_id', how='outer')
info_all = info_all.merge(contract, on='customer_id', how='outer')
```

**Observacio**   
Al compilar los data frames se reviso la existencia de valores nulos, se observo que existen varias columnas con valores nulos, estos valores no se pueden eliminar debido a que representan una gran cantidad de datos y eliminarlos podrian afectar el analisis , por esto se cambiaran los datos nulos por el  valor 'unknown'

```python
# funcion para remplazar los valores nulos
# para realizar esto se debe crear una ueva categotia para poder cambiarla  
def data_na(data,columns):
    
    
    for col in columns:
        
        #Nueva categoria
        category = 'unknown'

        # Agregar nueva categia  
        data[col] = data[col].cat.add_categories([category])
        # Rellenar valores nulos con la moda
        data[col] = data[col].fillna(category)
```

```python
col = ['internet_service','online_security', 'online_backup','device_protection','tech_support',
       'streaming_tv','streaming_movies','multiple_lines']


data_na(info_all,col)
# Revision de datos nulos 
null(info_all)
```

#### 3.2  Duración de contratos mensuales


```python
# Creacion de data para obtener el tiempo que los clientes utilizaron  el plan  por mes
time_month = info_all[(info_all['type'] == 'month-to-month')][['begin_date', 'end_date']]

# Calculamos los meses de duaración de los contratos y creacion de columna 'duration'
time_month['duration'] = round((time_month['end_date'] - time_month['begin_date']).dt.days / 30, 1)
```

```python
# Crear una figura con dos subplots: uno para el gráfico y otro para el texto
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5)) 

# Crear el histograma en el primer subplot
sns.histplot(time_month, x='duration', bins=20, ax=ax1)
ax1.set_title('Distribución de la Duración de los Contratos')
ax1.set_xlabel('Duración (mensual)')
ax1.set_ylabel('Cantidad de Contratos')

# Mostrar la salida de describe() 
describe_text = time_month['duration'].describe().to_string()
ax2.text(0.1, 0.5, describe_text, fontsize=12, va='center')
ax2.axis('off')  

# Ajustar el layout y mostrar la figura
plt.tight_layout()
plt.show()
```

### 4.5 Creacion de funcion de entrenamiento

```python

# Función para entrenar, evaluar y visualizar resultados
def evaluate_model(model, features_train, target_train, features_valid, target_valid, model_name,app_name):
    
    start_time_train = time.time()# inicio de tiempo de entrenamiento
    # Entrenar el modelo
    if isinstance(model, CatBoostClassifier):
        model.fit(features_train, target_train, verbose=50)  # CatBoost necesita verbose
    else:
        model.fit(features_train, target_train)
    
    time_train = round((time.time() -start_time_train),3)#tiempo de entrenamiento
    
    
    start_time_predic= time.time() # inicio de tiempo de prediccion
    # Predecir en el conjunto de validación
    predicted_valid = model.predict(features_valid)
    predicted_proba = model.predict_proba(features_valid)[:, 1]  # Probabilidades para la clase positiva
    
    time_predic = round((time.time() -start_time_predic),4)#tiempo de prediccion
    
    # Calcular la matriz de confusión
    mc = confusion_matrix(target_valid, predicted_valid)

    # Calcular métricas
    precision = round(precision_score(target_valid, predicted_valid),4)
    recall = round(recall_score(target_valid, predicted_valid), 4)
    f1 = round(f1_score(target_valid, predicted_valid), 4)
    auc_roc = round(roc_auc_score(target_valid, predicted_proba), 4)
    #Titulo del modelo en ejecucion
    
    fig, ax = plt.subplots(1,figsize=(12,.1))
    ax.axis('off') # Ocultar los ejes (opcional, para que solo se vea la tabla)
    ax.set_title(f"{app_name} Model: {model_name}", fontsize=20, fontweight='bold',loc='center')
    plt.show()
    
    # Crear una figura con dos subplots: uno para el gráfico y otro para el texto
    fig, ( ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))

   
    # Graficar la matriz de confusión en el primer subplot
    disp = ConfusionMatrixDisplay(confusion_matrix=mc)
    disp.plot(ax=ax1)
    ax1.set_title(f'Confusion Matrix - {model_name}')

    # Calcular la curva ROC
    fpr, tpr, thresholds = roc_curve(target_valid, predicted_proba)

    # Graficar la curva ROC en el segundo subplot
    ax2.plot(fpr, tpr, label=f'AUC-ROC = {auc_roc:.4f}')
    ax2.plot([0, 1], [0, 1], 'k--')  # Línea de referencia para un modelo aleatorio
    ax2.set_xlabel('Tasa de Falsos Positivos (FPR)')
    ax2.set_ylabel('Tasa de Verdaderos Positivos (TPR)')
    ax2.set_title(f'Curva ROC - {model_name}')
    ax2.legend(loc='lower right')

    # Ajustar el layout y mostrar la figura
    plt.tight_layout()
    plt.show()
    
    # Imprimir métricas
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"AUC-ROC: {auc_roc:.4f}")
    
    # Calcular el tiempo de ejecución

    print(f"Tiempo de ejecución del entrenamiento : {time_train:.4f} segundos")
    print(f"Tiempo de ejecución de la prediccion: {time_predic:.4f} segundos")
    
    return precision,recall,f1,auc_roc,time_train,time_predic
    
```

### 4.8 Undersampling

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)]+ [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] 
        + [target_ones])

    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345
    )

    return features_downsampled, target_downsampled


features_downsampled, target_downsampled = downsample(features_train, target_train, 0.3)
```
